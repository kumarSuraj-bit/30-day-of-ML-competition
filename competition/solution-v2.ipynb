```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
```

Data Fetch

```python
raw_test_data = pd.read_csv('test.csv')
raw_train_data = pd.read_csv('train.csv')
```

Preparing Data for Traing and Testing

```python
y = raw_train_data['target']
```

```python
raw_train_data = raw_train_data.drop('id',axis=1)
raw_train_data = raw_train_data.drop('target',axis=1)
```

```python
train_cat = raw_train_data.select_dtypes(object)
test_cat = raw_test_data.select_dtypes(object)
```

changing category data to numerical data

```python
for col in train_cat:
    le = LabelEncoder()
    raw_train_data[col] = le.fit_transform(raw_train_data[col])
```

```python
train = raw_train_data.copy()
```

```python
raw_train_data.shape  ,train.shape
```

```python
train.head(10)
```

Model and training

```python
from sklearn.model_selection import train_test_split
import xgboost as XGB
```

```python
x_train,x_test,y_train,y_test = train_test_split(train,y,random_state=64)
```

```python
model = XGB.XGBRegressor(colsample_bytree=0.4603, gamma=0.0468, 
                             learning_rate=0.05, max_depth=3, 
                             min_child_weight=1.7817, n_estimators=2200,
                             reg_alpha=0.4640, reg_lambda=0.8571,
                             subsample=0.5213, random_state =7, nthread = -1)
```

```python
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

Prepare Test data

```python
raw_test_data.head(10)
```

```python
test_id = raw_test_data['id']
```

```python
raw_test_data  = raw_test_data.drop('id',axis=1)
```

```python
for col in test_cat:
    le = LabelEncoder()
    raw_test_data[col] = le.fit_transform(raw_test_data[col])
```

```python
test = raw_test_data.copy()
```

```python
raw_test_data.shape,test.shape
```

Save Results in a File

```python
y_predict = model.predict(test)

my_file = pd.DataFrame()

my_file['id'] = test_id
my_file['target'] = y_predict
```

```python
my_file.to_csv('solution_s_v2.csv',index=False)
```

```python
from sklearn.metrics import mean_absolute_error

cal_predict = model.predict(x_test)

print("Mean Absolute Error: " + str(mean_absolute_error(cal_predict, y_test)))
```
